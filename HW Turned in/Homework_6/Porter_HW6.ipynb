### Definition of Reversibility


In class, we defined a reversible system as one whose dynamics looked the same whether time runs forward or backwards. Strogatz says that any second-order system that is invariant under $t\to-t,~y\to-y$ is reversible. A reversible system of this type displays the following behavior in the Euclidian coordinate system:

$$
\begin{array}{c}
     \dot{x}=f(x,y)\\
     \dot{y}=g(x,y)
 \end{array} \longrightarrow \begin{array}{c}
     t\rightarrow-t,~~~y\rightarrow-y\\ 
 \end{array} \longrightarrow \begin{array}{c}
     \dot{x}=-f(x,-y)\\
     \dot{y}=g(x,-y)
 \end{array}
$$

Here, $\dot{x}$ is odd in $y$ and $\dot{y}$ is even in $y$. If we say $G$ is this transformation, we can  think of $G$ as an operation on some system $(\dot{x},\dot{y})$, and the composition $G\circ G=I$ is equivalent to $G(G(\dot{x},\dot{y}))=(\dot{x},\dot{y})$. This is essentially the definition of the inverse (for systems where the inverse exists). We can see that performing the $G$ operation $t\to-t,~y\to-y$ again will give us our original function in return:

$$
\begin{array}{c}
     \dot{x}=-f(x,-y)\\
     \dot{y}=g(x,-y)
 \end{array} \longrightarrow \begin{array}{c}
     t\rightarrow-t,~~~y\rightarrow-y\\ 
 \end{array} \longrightarrow \begin{array}{c}
     \dot{x}=f(x,y)\\
     \dot{y}=g(x,y)
 \end{array}
$$

This aligns with the definition we discussed in class, which uses this transformation to show how reversible systems will have mirrored phase paths going opposite directions. In the example we just discussed, the paths are identical on either side of the $x$-axis, due to the even/odd properties conditions mentioned.




### Other Involutions


Our definition of reversibility primarily depends on the results of the transformation $t\to-t$, but we can make transformations other than $y\to-y$ under certain conditions. As long as fixed points and circular orbits are symmetrical across some line in the phase plane, the system can be reversed. For example, for the transformation $t\to-t, ~x\to-x$:

$$
\begin{array}{c}
     \dot{x}=f(x,y)\\
     \dot{y}=g(x,y)
 \end{array} \longrightarrow \begin{array}{c}
     t\rightarrow-t,~~~x\rightarrow -x\\ 
 \end{array} \longrightarrow \begin{array}{c}
     \dot{x}=-f(-y,-x)\\
     \dot{y}=g(-y,-x)
 \end{array}
$$

This meets reversibility criteria because $\dot{x}$ is odd in $x$ and $\dot{y}$ is even in $x$. This results in paths mirroring on opposite sided of the $y$-axis. Another example is $t\to-t, ~x\to-x, ~y\to-y$:

$$
\begin{array}{c}
     \dot{x}=f(x,y)\\
     \dot{y}=g(x,y)
 \end{array} \longrightarrow \begin{array}{c}
     t\rightarrow-t,~~~x\rightarrow-x,~~~y\rightarrow-y\\ 
 \end{array} \longrightarrow \begin{array}{c}
     \dot{x}=-f(-x,-y)\\
     \dot{y}=-g(-x,-y)
 \end{array}
$$

Here, $\dot{x}$ is even in $x$ and odd in $y$, and $\dot{y}$ is even in $y$ and odd in $x$. The opposite could also be true: $\dot{x}$ is even in $y$ and odd in $x$, and $\dot{y}$ is even in $x$ and odd in $y$. This system would have both reflectional symmetry across both the $x$ and $y$ axes 

### Show That a Particular Fixed Point is a Center

Theorem 6.6.1 tells us that in reversible system with a linear center, all trajectories near the center are closed orbits. This indicates that small perturbations, such as those that may arise due to nonlinear terms, do not degrade to spirals. We can then say that a linear center in a reversible system is robust, and therefore is also a nonlinear center. 

To explain a bit more fully: we first assume that the system is reversible. If there is a fixed point at the origin with purely imaginary eigenvalues, then we can deduce two things:

1) Any orbit traveling in the first and fourth quadrant must be reflected in the second and third quadrant. 

2) The lack of real parts of the eigenvalues indicate that there is no growth or decay in the system.

For the first example, we had $x\to-x$. This involution reflects points across the $y$-axis. If there is a fixed point at the origin, it cannot decay or grow into a spiral, since that would violate the closed orbit rule. Since the eigenvalues are purely imaginary, there is no real value to compare growth or decay near that point, so it must be a linear center at the very least. If it is a linear center and all orbits near it are closed, then it must be a nonlinear center. The exact same argument applies to the involution $x\to-x$ and $y\to-y$. In this case, the symmetry would be both reflective and rotational.

## Problem 3 (6.6.8)

Consider the system 

$$ \dot{x}=\frac{\sqrt{2}}{4}x(x-1)\sin\phi, ~~~~\dot{\phi}=\frac{1}{2}\left[\beta-\frac{1}{\sqrt{2}}\cos\phi-\frac{1}{8\sqrt{2}}x\cos\phi\right] $$

where $0\leq x\leq1$ and $-\pi\leq\phi\leq\pi$. Since the system is $2\pi$-periodic in $\phi$, it may be considered as a vector field on a _cylinder_. The $x$-axis runs along the cylinder, and the $\phi$-axis wraps around it. Note that the cylindrical phase space is finite, with edges given by the circles $x=0$ and $x=1$. 

### (a) Show that the system is reversible

Say $\dot{x}=f(x,\phi)$ and $\dot{\phi}=g(x,\phi)$. We can see that since $f$ is a function of $\sin{\phi}$, it is odd in $\phi$ so $f(x,-\phi)=-f(x,\phi)$. Also, $g$ is even in $\phi$, so $g(x,-\phi)=g(x,\phi)$. Therefore, if we make the transformations $t\to-t$ and $\phi\to-\phi$, the dynamics remain the same with time reversal. The phase portrait should look symmetric across the $x$-axis. 

### (b) Verify that for $\frac{9}{8\sqrt{2}}>\beta>\frac{1}{\sqrt{2}}$, the system has three fixed points on the cylinder, one of which is a saddle. Show that this saddle is connected to itself by a homoclinic orbit that winds around the waist of the cylinder. Using reversibility, prove that there is a _band of closed orbits_ sandwiched between the circle $x=0$ and the homoclinic orbit. Sketch a phase portrait of the cylinder, and check your results by numerical integration.



Setting $\dot{x}$ and $\dot{\phi}$ equal to zero, we get several possible fixed points. From $\dot{x}$, we can easily see that zeros occur for $x=0,~ x=1, ~\phi=-\pi, ~\phi=0$, and $\phi=\pi$. Plugging these values into $\dot{\phi}$ one at a time and solving for $\dot{\phi}=0$ yields the following fixed points:

$
\left(0,\cos^{-1}\left(\sqrt{2}\beta\right)\right),
~~~\left(1,\cos^{-1}\left(\frac{8\sqrt{2}}{9}\beta\right)\right), \\
(-8(\sqrt{2}\beta+1),0) , 
~~~(8(\sqrt{2}\beta-1),\pm\pi) 
$

Since the first two fixed points are not defined for $\frac{1}{\sqrt{2}}<|\beta|<\frac{9}{8\sqrt{2}}$, there are only three fixed points on the interval $\frac{1}{\sqrt{2}}<\beta<\frac{9}{8\sqrt{2}}$. 



We can quickly find the saddle by examining a phase portrait and the nullclines. The nullclines for $\dot{x}$ are simply straight lines where the zeros occur. For $\dot{\phi}$, we can take $x$ as a function of $\phi$, which gives us $x=8(\sqrt{2}\beta\sec\phi-1)$. Let's assume that $\beta$ is within our interval in question and equal to 0.75. Below is the phase portrait for the system, with the $\dot{x}$ nullclines in blue, the $\dot{\phi}$ nullclines in green, and the fixed points in black.

```python
beta = 0.75
pi = np.pi

def f(Y, t):
    xvals, pvals = Y
    xdot = np.sqrt(2.)/4.*xvals*(xvals-1.)*np.sin(pvals)
    phidot = 1./2.*(beta-np.cos(pvals)/np.sqrt(2.)*(1+xvals/8.))
    return [xdot,phidot]
        
xvals = np.linspace(0,1.1,25)
pvals = np.linspace(-pi,pi,16)
Y1, Y2 = np.meshgrid(xvals, pvals)
u, v = np.zeros(Y1.shape), np.zeros(Y2.shape)
NI, NJ = Y1.shape
t = 0
tspan = np.linspace(0, 150, 500)

for i in range(NI):
    for j in range(NJ):
        x = Y1[i, j]
        y = Y2[i, j]
        yprime = f([x, y], t)
        u[i,j] = yprime[0]/np.sqrt(yprime[0]**2+yprime[1]**2)
        v[i,j] = yprime[1]/np.sqrt(yprime[0]**2+yprime[1]**2)


Q = plt.quiver(Y1, Y2, u, v, color='r')
plt.plot([0,1.1],[0,0],color='b')

pvals2 = np.linspace(-1,1,50) 
plt.plot(8.*(np.sqrt(2.)*beta*1./np.cos(pvals2)-1.),pvals2,color='g')

plt.plot(8.*(np.sqrt(2)*beta*1./np.cos(0)-1.),0,'o',color='k')
plt.plot(1.,np.arccos(beta/(1./np.sqrt(2)+1./(8.*np.sqrt(2))*1.)),'o',color='k')
plt.plot(1.,-np.arccos(beta/(1./np.sqrt(2)+1./(8.*np.sqrt(2))*1.)),'o',color='k')
plt.plot([1,1],[-pi,pi],'b')


for y0 in [[.483,.009],[.483,-2.*pi+.009]]:    
    ys = odeint(f, y0, tspan)
    plt.plot(ys[:,0], ys[:,1], 'k-')
           
for y1 in np.linspace(.001,.009,4):
    for y2 in [-pi]:
        ys = odeint(f, [y1,y2], tspan)
        plt.plot(ys[:,0], ys[:,1])

plt.xlim([0,1.05])
plt.ylim([-pi,pi])

plt.xlabel('$x$')
plt.ylabel('$\phi$')
print 'Problem 6.6.8'
```

We can see that the saddle is at the fixed point along the $\phi=0$ line, and is unstable for positive $\phi$ and stable for negative $\phi$. Likewise, the fixed point for positive $\phi$ is unstable and the fixed point for $-\phi$ is stable. To show that this saddle in connected to itself by a homoclinic orbit that winds around the cylinder, we look at the behavior of the manifolds near the saddle point. To do this, we can choose a point very close to the saddle along the unstable eigenvector.

We can verify that the fixed point where $\phi=0$ is a saddle by taking the Jacobian matrix of the system, plugging in the fixed point, and finding the eigenvalues.

$$ J= \left(\begin{array}{cc}
(2x-1)\left(\frac{\sqrt{2}}{4}\right)\sin\phi & x^2-x \\
\left(\frac{1}{16\sqrt{2}}\right)\cos\phi & \frac{1}{2\sqrt{2}}-\frac{1}{16\sqrt{2}}x 
\end{array}\right) $$

Using this as our starting point, we can see that a plot of this trajectory moves away from the saddle point and toward the $\phi$-axis, coming near $x=0$. Once it crosses the nullcline at $\phi=\pi$, it starts moving toward the (next) saddle point. We can see this behavior at the bottom of the plot, where the trajectory from the saddle on $\phi=-2\pi$ approaches the saddle on $\phi=0$. The line $x=0$ is also a trajectory, and when imagining this behavior around a cylinder, we have a closed circular orbit on the $x=0$ nullcline that is "inside" our homoclinic orbit. 


The Poincare-Bendixon theorem tells us that if a trajectory is inside another trajectory that is closed and bounded, and there is no fixed point _within_ this other trajectory's region, then the trajectory must eventually approach a closed orbit. Our $x=0$ nullcline and homoclinic orbit meet this criteria. And since the system is reversible in $\phi$, all  paths will be reflected across the $x$-axis. Any path that emerges from the $\phi=-\pi$ nullcline must return to the same $x$-value at $\phi=\pi$. Therefore, any trajectory that begins on the $x$-axis between 0 and the saddle point must eventually close around the cylinder.
