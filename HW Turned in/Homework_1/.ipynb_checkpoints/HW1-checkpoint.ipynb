## Problem 1

#### a) Solve and plot for $t\geq 0$.

$$
\dot{x} + 3x = \cos(3t), ~ x(0) = 2
$$

The above differential equation has the form

$$
\dfrac{dx}{dt}+p(t)x(t)=q(t)
$$

and can therefore be solved using an integrating factor. Since $p(t) = 3$, we define the integrating factor $v(t)$ as 

$$
v(t) = e^{\int_{0}^{t} 3 du} = e^{3t}.
$$

Multiplying the differential equation by the integrating factor yields

$$
\dfrac{d}{dt}[x(t)e^{3t}]=cos(3t)e^{3t} \hspace{3mm} \longrightarrow \hspace{3mm} x(t)=e^{-3t}\int_{0}^{t}cos(3t)e^{3t}dt+Ce^{-3t}.
$$

For the integral, we use the integration by parts technique twice:

$$
\begin{align}
\int_{0}^{t}\cos(3s)e^{3s}ds 
        & = \frac{1}{3}cos(3t)e^{3t}+\int_{0}^{t}\sin(3s)e^{3s}ds \\
        & = \frac{1}{3}cos(3t)e^{3t}+\frac{1}{3}sin(3t)e^{3t}-\int_{0}^{t}\cos(3s)e^{3s}ds \\
        & = \frac{1}{6}e^{3t}(cos(3t)+sin(3t))
\end{align}
$$

Substituting this value back into $x(t)$ and using the initial value condition, we get

$$
x(t) = \frac{1}{6}(cos(3t)+sin(3t))+\frac{11}{6}e^{-3t}.
$$

We can now plot the solution to $x(t)$ for $0\leq t\leq 10$. Note that the solution continues to oscillate without significant period or amplitude change for $10 \leq t$, as the exponential term approaches zero. We can then say that for $t>10$,

$$
x(t)\approx\frac{1}{6}(cos(3t)+sin(3t))
$$

```python
tvals = np.linspace(0,10,100)
xvals = 1./6.*(ncos(3*tvals) + nsin(3*tvals)+11.*ne(-3.*tvals)) 

plt.plot(tvals,xvals)
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.title('$x(t)$ vs. $t$')

print 'Problem 1a'
```

#### b) Solve and plot for $t\geq 1$.

$$
\dot{x} + \frac{x}{t} = \sin(2t), ~ x(1) = 3
$$

This differential equation has the same form as the previous one, with $p(t)=t^{-1}$, $q(t)=sin(2t)$, and an integrating factor $v(t)=t$. We can then follow the same steps (again using integration by parts) to reach a solution:

$$
\frac{d}{dt}[t\cdot x(t)]=t\cdot sin(2t) \\
\begin{align}
x(t) & =\frac{1}{t}\int_0^ts\cdot sin(2s)ds-\frac{c_1}{t} \\
     & = \frac{1}{t}\left(-\frac{1}{2}t\cdot cos(2t)+\frac{1}{4}sin(2t)\right)-\frac{c_1}{t} \\
     & = -\frac{1}{2}cos(2t)+\frac{1}{4t}sin(2t)-\frac{sin(2)-2cos(2)-12}{4t}
\end{align}
$$

Now we can plot the solution for $1\leq x$. Again, the solution continues to oscillate for larger values of $t$. For very large $t$, the fractional term approaches zero. We can then say that

$$
x(t)\approx-\frac{1}{2}cos(2t)+\frac{1}{4t}sin(2t)
$$

for $t$ >> 10.

```python
tvals = np.linspace(1,14,100)
xvals = -1./2.*ncos(2.*tvals) + 1./(4.*tvals)*nsin(2.*tvals)-(nsin(2)-2.*ncos(2)-12.)/(4.*tvals)

plt.plot(tvals,xvals)
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.title('$x(t)$ vs. $t$')

print 'Problem 1b'
```

$$
\begin{align}
  \dddot{x} & = cos(x)[(cos(x)sin(x)]+sin(x)[-sin(x)sin(x)]\\

\end{align}
$$

Now our roots are $x=n\pi$, $x=\frac{\pi}{4}+n\pi$, and $\frac{3\pi}{4}+n\pi$. Since only one of these results in $\ddot{x}>0$, we know that the flow has a maximum positive acceleration at $x=\frac{\pi}{4}+n\pi$.

## Problem 3 (2.2.3, 2.2.4, 2.2.7)

Analyze the following equations graphically. In each case, sketch the vector field on the real line, find all the fixed points, classify their stability, and sketch the graph of $x(t)$ for different initial conditions. 

#### 2.2.3) $\dot{x}=x-x^3$

To find the fixed points, we set $\dot{x}=0$ and solve. This yields $x=0,~x=\pm1$. Since $\dot{x}$ flows from positive to negative at $x=-1$, we know that $x=-1$ is a stable fixed point. It follows that $x=0$ is unstable and $x=1$ is stable (the vector field plot below confirms this). We can find the analytical solution to this equation by using partial fraction decomposition and using $C$ as any arbitrary constant:

$$
\frac{dx}{dt}=x-x^3 ~~~ \longrightarrow ~~~ \frac{dx}{x(1-x)(1+x)}=dt \\
\begin{align}
t & =\int\left(\frac{1}{x}-\frac{1}{2(x-1)}-\frac{1}{2(x+1)}\right)dx +C \\
    & =ln|x|-\frac{1}{2}ln|x-1|-\frac{1}{2}ln|x+1| + C
\end{align}
$$

Solving for $x(t)$, we find

$$
e^t=Cx(x-1)^{-1/2}(x+1)^{-1/2} \\
(x^2-1)e^{2t}=Cx^2 \\
x^2e^{2t}-Cx^2=e^{2t} \\
x(t) = \pm\frac{e^t}{\sqrt{C+e^{2t}}}
$$

The following graphs show a plot of the vector field for $\dot{x}$ and plot of the analytical solution $x(t)$ with varying values of $C$ to represent different initial conditions. Note that the fixed points $x(t)=0$ and $x(t)=1$ are shown in dotted lines for the solution plot. For better readability, this plot omitted the solutions near the fixed point $x=-1$, however, it is simply a mirror image of the shown plot about the asymptote $x(t)=0$.


```python
xvals = np.linspace(-1.5,1.5,30)
yvals = np.zeros(xvals.size)

xprime = xvals-xvals**3
yprime = np.zeros(xvals.size)

plt.plot(xvals,xvals-xvals**3,ls='--',color='k')
plt.quiver(xvals,yvals,xprime,yprime,color='r')
plt.xlabel('$x$')
plt.ylabel('$dx/dt$')
plt.title('Vector Field for $\dot{x}$')

print 'Problem 2.2.3'
```

#### 2.2.4) $\dot{x}=e^{-x}sin(x)$

To find the fixed points of $\dot{x}$, we need only examine values of $x$ where $sin(x)=0$. The fixed points are $x=n\pi$, with $n=0,\pm1,\pm2,...$ and have alternating stability. The first vector field plot below illustrates this alternating stability, with stable points occurring at $x=(2n+1)\pi$ and the unstable points at $x=2n\pi$. Only the interval $0<x<2\pi$ is shown, as negative values of $x$ have very large amplitudes and values of $x$ larger than $2\pi$ have very small amplitudes. There is no nice analytic solution to this differential equation, so a numerical ODE solver was used for the second solution plot below.


```python
xvals = np.linspace(0,2.*Pi,90)
yvals = np.zeros(xvals.size)
xprime = ne(-xvals)*nsin(xvals)
yprime = np.zeros(xvals.size)

plt.plot(xvals,ne(-xvals)*nsin(xvals),ls='--',color='k')
plt.quiver(xvals,yvals,xprime,yprime,color='r')
plt.xlabel('$x$')
plt.ylabel('$dx/dt$')
plt.title('Vector Field for $\dot{x}$')

print 'Problem 2.2.4'
```

```python
def dx_dt(x,t):
    return ne(-x)*nsin(x)
minv = -10
maxv = 100
stp = Pi/6.
tvals = np.linspace(minv,maxv,1000)
x0 = [stp*1,stp*2,stp*3,stp*4,stp*5,stp*7,stp*8,stp*9,stp*10,stp*11]
fp = [0,Pi,2*Pi]

for x in x0:
    xt = spint.odeint(dx_dt,x,tvals)
    xt = np.array(xt).flatten()
    plt.plot(tvals,xt)

for x in fp:
    plt.plot([minv,maxv],[x,x],ls='--',color='k')
    
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.title('$x(t)$ vs. $t$')

print 'Problem 2.2.4'
```

#### 2.2.7) $\dot{x}=e^x-cos(x)$

This differential equation does not have fixed points that can be solved explicitly. We can, however, determine the general behavior and stability of $\dot{x}$ by examining the vector field plot:

```python
xvals = np.linspace(-10,1.2,30)
yvals = np.zeros(xvals.size)
xprime =  ne(xvals)-ncos(xvals)
yprime = np.zeros(xvals.size)

plt.plot(xvals,ne(xvals)-ncos(xvals),ls='--',color='k')
plt.quiver(xvals,yvals,xprime,yprime,color='r')
plt.xlabel('$x$')
plt.ylabel('$dx/dt$')
plt.title('Vector Field for $\dot{x}$')

print 'Problem 2.2.7'
```

Here we can see that the flow falls into a sinusoidal pattern as $x$ decreases, due to the dwindling influence of the exponential term. For greater negative values of $x$, we can say that $\dot{x}\approx-cos(x)$. This indicates that the fixed points occur at approximately $x=(2n+1)\pi/2$, with $n$=0,-1,-2,... for larger negative values of $x$. In addition, the stability of the fixed points are as we would expect, with stable equilibria where $\dot{x}$ switches from positive to negative, and unstable equilibria for the alternative case. While no analytical solution is possible, we can observe a family of solutions and their behavior in relation to the approximate fixed points for $-5\leq x(t) \leq 0$:

```python
def dx_dt(x,t):
    return ne(x)-ncos(x)
minv = 0.
maxv = 2.
stp = Pi/6.
tvals = np.linspace(minv,maxv,100)
x0 = [-.431,-.862,-1.7215,-2.15,-2.5785,-3.007,-3.4355,-3.864,-4.2925]
fp = [0,-1.293,-4.721]

for x in x0:
    xt = spint.odeint(dx_dt,x,tvals)
    xt = np.array(xt).flatten()
    plt.plot(tvals,xt)

for x in fp:
    plt.plot([minv,maxv],[x,x],ls='--',color='k')
    
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.title('$x(t)$ vs. $t$')

print 'Problem 2.2.7'
```

## Problem 4 (2.2.13)

The velocity $v(t)$ of a skydiver falling to the ground is governed by $m\dot{v}=mg-kv^2$, where $m$ is the mass of the skydiver, $g$ is the acceleration due to gravity, and $k \gt 0$ is a constant related to the amount of air resistance.

#### a) Obtain the analytical solution for $v(t)$, assuming that $v(0)=0$.

Starting with $m\dot{v}=mg-kv^2$, we can use partial fraction differentiation and hyperbolic trigonometric identities to solve the differential equation. Again, $C$ will signify some arbitrary constant.

$$
m\frac{dv}{dt}=mg-kv^2 ~~~ \longrightarrow ~~~ m\frac{dv}{mg-kv^2}=dt \\
\begin{align}
    t & = m\int\frac{dv}{mg-kv^2}+C \\
      & = \frac{1}{g} \int\frac{dv}{1-\frac{k}{mg}v^2}+C \\
      & = \sqrt{\frac{m}{kg}}\int\frac{1}{2}\left(\frac{1}{1+v\sqrt{\frac{k}{mg}}}-\frac{1}{1-v\sqrt{\frac{k}{mg}}}\right)dt+C \\
      & = \frac{1}{2}\sqrt{\frac{m}{kg}}\left(ln\left|1+v\sqrt{\frac{k}{mg}}\right|-ln\left|1-v\sqrt{\frac{k}{mg}}\right|\right) +C\\
      & = \sqrt{\frac{m}{kg}}arctanh\left(v\sqrt{\frac{k}{mg}}\right) +C
\end{align}
$$
So then
$$
v\sqrt{\frac{k}{mg}}=tanh\left(t\sqrt{\frac{kg}{m}}+C\right) \\
v = \sqrt{\frac{mg}{k}}tanh\left(t\sqrt{\frac{kg}{m}}+C\right) \\
$$

Since we know that $v(0)=0$, Then $C$ must also be zero, giving us:
 
$$
 v(t) = \sqrt{\frac{mg}{k}}tanh\left(t\sqrt{\frac{kg}{m}}\right)
$$

#### b) Find the limit of $v(t)$ as $t \rightarrow \infty$. This limiting velocity is called the $terminal~velocity$. 

As $t \rightarrow \infty$, $tanh\left(t\sqrt{\frac{kg}{m}}\right)\rightarrow 1$, leaving the terminal velocity as simply:

$$
v_T=\sqrt{\frac{mg}{k}}
$$

#### c) Give a graphical analysis of this problem, and thereby re-derive a formula for the terminal velocity.

Examining our original differential equation, $m\dot{v}=mg-kv^2$, we can look at the differential equation

$$
\frac{dv}{dt}=g-\frac{k}{m}v^2
$$

Even without specific parameters, we can still get a general idea of the behavior of this system. We know that there are two fixed points, one stable at $\sqrt{\frac{gm}{k}}$ (representing terminal velocity) and one unstable at $-\sqrt{\frac{gm}{k}}$. We also know that this is a simple downward parabola shifted vertically. If we use simple values for the unknown parameters, say $g=10$, $m=70$, and $k=3$, we can visualize a phase plot as shown below. Note that the positive root represents the terminal velocity.



```python
xvals = np.linspace(-20,20,25)
yvals = np.zeros(xvals.size)
xprime = 10.-(3./70.)*xvals**2.
yprime = np.zeros(xvals.size)

plt.plot(xvals,10.-(3./70.)*xvals**2.,ls='--',color='k')
plt.quiver(xvals,yvals,xprime,yprime,color='r')
plt.xlabel('$v$')
plt.ylabel('$dv/dt$')
plt.title('Vector Field for $\dot{v}$')

print 'Problem 2.2.13'
```

If we want to visualize some general solutions for this equation, we can give generic values for $g$ and $\frac{k}{m}$, resulting in an equation that might look like $\frac{dv}{dt}=A-Bv^2$, with $A=g$ and $B=\frac{k}{m}$. Since the derivation of terminal velocity is largely the same as in parts a) and b), it will be omitted here. The result looks like:

$$v(t)=\sqrt{\frac{A}{B}}tanh\left(t\sqrt{AB}\right) ~~~\longrightarrow ~~~ V_T=\sqrt{\frac{A}{B}}$$

By simplifying the equation, we can easily examine some solution plots: one where $A$ is varied, and one where $B$ is varied.

```python
A0 = [8.,9.,10.,11.,12.]
B = .02856

minv = 0
maxv = 10

tvals = np.linspace(minv,maxv,100)

for A in A0:
    xvals = np.sqrt(A/B)*np.tanh(tvals*np.sqrt(A*B))
    plt.plot(tvals,xvals)
    
plt.xlabel('$t$')
plt.ylabel('$v(t)$')
plt.title('$v(t)$ vs. $t$')

print 'Problem 2.2.13, varied A'
```

```python
A = 10.
B0 = [0.017136, 0.022848,0.02856,0.034272,0.039984]

minv = 0
maxv = 10

tvals = np.linspace(minv,maxv,100)

for B in B0:
    xvals = np.sqrt(A/B)*np.tanh(tvals*np.sqrt(A*B))
    plt.plot(tvals,xvals)
    
plt.xlabel('$t$')
plt.ylabel('$v(t)$')
plt.title('$v(t)$ vs. $t$')

print('Problem 2.2.13, varied B')
```

As one might expect, increasing $A$ results in reaching a greater terminal velocity more quickly and therefore producing a greater value for the stable fixed point of the system. This is of course logical, as increasing the gravity force acting on the mass would cause a much faster freefall. Increasing $B$, which could correspond to either increasing $k$ or decreasing $m$, slows the freefall and results in a lower terminal velocity. Again this is logical, as increased air resistance will slow a falling body, and decreased mass will in turn decrease the weight of the body.

Lets say that $\dot{x}=v(t)$, where $x(t)$ is the position equation. This gives us:

$$
\frac{dx}{dt}=\sqrt{\frac{mg}{k}}tanh\left(t\sqrt{\frac{kg}{m}}\right) \\
\begin{align}
x(t) & =\sqrt{\frac{mg}{k}}\int tanh\left(t\sqrt{\frac{kg}{m}}\right) +C \\

$$

$$

$$
